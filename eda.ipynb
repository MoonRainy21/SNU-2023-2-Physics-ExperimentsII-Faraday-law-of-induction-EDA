```python
import pandas as pd
import matplotlib.pyplot as plt

# Mac font path
fontpath = '/Library/Fonts/NanumGothic.ttf'
plt.rcParams['axes.unicode_minus'] = False  

filename = "직류 중간 8.5 1.88.csv"
def plot(filename_list, show=True, save=False, legend=[], title=None):
  if type(filename_list) == str:
    filename_list = [filename_list]
  for filename in filename_list:
    df = pd.read_csv(filename)
    plt.plot(df['V'].values)
    plt.grid()
  plt.legend(legend)
  plt.xlabel('Steps')
  plt.ylabel('Voltage')
  plt.title(title)
  if show: plt.show()
  if save: plt.savefig(filename[:-4]+'.png')
  plt.close()
```

```python
plot(["7.0 1.80.csv","8.5 1.89.csv"], legend=['7.0V', '8.5V'], title='7.0V vs 8.5V')
```

```python
# 푸리에 변환한 결과로 그래프 그리기
import numpy as np

def plot_fft(filename_list, show=True, save=False, legend=[], title=None):
  if type(filename_list) == str:
    filename_list = [filename_list]
  for filename in filename_list:
    df = pd.read_csv(filename)
    fouier = np.fft.fft(df['V'].values) / len(df['V'].values) * 2
    plt.plot(np.abs(fouier[:100]))
    plt.grid()
  plt.legend(legend)
  plt.xlabel('Frequency')
  plt.ylabel('Voltage')
  plt.title(title)
  if show: plt.show()
  if save: plt.savefig(filename[:-4]+'.png')
  plt.close()
```

```python
plot_fft(["7.0 1.80.csv","8.5 1.89.csv"], legend=['7.0V', '8.5V'], title='7.0V vs 8.5V (FFT)')
```

```python
def get_freqency(filename):
  if type(filename) == str:
    filename = [filename]
  ans = []
  for filename in filename:
    df = pd.read_csv(filename)
    fouier = np.fft.fft(df['V'].values) / len(df['V'].values) * 2
    ans.append([np.argmax(np.abs(fouier[:100])), np.max(np.abs(fouier[:100]))])
  return ans

ans = get_freqency(["7.0 1.80.csv","8.5 1.89.csv"])
for i in ans:
  print(i, i[0] / i[1])
```
